## 2. HDI Value vs Total Medal test set

### Hypothesis - The higher the HDI value, the more awards won

```python
# Creating a plot to see the two variables together
df.plot(x = 'HDI_value', y='tot_medal',style='o')
plt.title('HDI_value vs Total Medals')  
plt.xlabel('HDI_value')  
plt.ylabel('Total Medals')  
plt.savefig('HDI vs Tot_Med scat.png')
plt.show()
```

```python
# reshaping values
X = df['HDI_value'].values.reshape(-1,1)
y = df['tot_medal'].values.reshape(-1,1)
```

```python
#Checking X array
X
```

```python
# Checking y array
y
```

```python
# Creating training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Creating a regression object
regression = LinearRegression()
```

```python
# Fitting the regression to the training set
regression.fit(X_train, y_train)
```

```python
# Predicting values of y by using X
y_predicted = regression.predict(X_test)
```

```python
# Viewing plot with regression line in it
plot_test = plt
plot_test.scatter(X_test, y_test, color='blue', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('HDI Value vs Total Medals (Test set)')
plot_test.xlabel('HDI Value')
plot_test.ylabel('Total Medals')
plt.savefig('HDI_value vs Tot_Med (test set).png')
plot_test.show()
```

```python
# Creating objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)
```

```python
# Checking summary stats
print('Slope:', regression.coef_)
print('Mean squared error:', rmse)
print('R2 score:', r2)
```

#### Notes
 - The Slope reads a positive relationship. This means that the higher the HDI value the more medals are won. The graph also reflects this but it is clear that the line doesn't cover a large portion of the data points which weakens the result. 
 - The MSE is very large. This indicates that the line is far from reflecting all of the data points. 
 - A R2 score of 0.087 proves that this model is unsuitable for predictions. 

Although the relationship is a positive, the summary stats clearly indicate that this model is not a good fit and shouldn't be used to predict medal numnbers.

```python
# Checking dataframe comparing the real and the predicted values of y
data = pd.DataFrame({'Actual' : y_test.flatten(), 'Predicted' : y_predicted.flatten()})
data.head(30)
```

### Notes
 - The Actual and Predicted colmuns are very different from each other. This clearly shows how unsuitable the model is. 

## 3. HDI and Total Medal Training set

```python
# Predicting against the training set
y_predicted_train = regression.predict(X_train)
```

```python
plot_test = plt
plot_test.scatter(X_train, y_train, color='green', s = 15)
plot_test.plot(X_train, y_predicted_train, color='red', linewidth =3)
plot_test.title( 'HDI Value vs Total Medals (Training set)')
plot_test.xlabel('HDI Value')
plot_test.ylabel('Total Medals')
plt.savefig('HDI_value vs Tot_Med (training set).png')
plot_test.show()
```

### Notes 
 - The Slope and the MSE score have risen slightly while the R2 score has decreased slightly. The R2 decreasing is a bad sign as the closer it is to 0 the higher the variance between the data points. However, the increase in the Slope and MSE indicate that the positive relationship (as shown from the slope) is stronger and there is last deviation from the line. This is also seen in the graph as the points are clustered more closely around the line.
