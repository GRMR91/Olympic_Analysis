## 4. Data grouping and further analysis

### Grouping medals

```python
# Inspecting the tot_medal column to see for it should be grouped for analysis
mast_df['tot_medal'].describe()
```

```python
# Creating loop to catorgerise medal tiers
medal_result = []
counter = 0
highVal = 0
for value in mast_df['tot_medal']:
    if value == 0:
        medal_result.append('No medals')
    elif value > 0 and value <= 25:
        medal_result.append('1-25')
    elif value > 25 and value <=50:
        medal_result.append('26-50')
    elif value > 50 and value <=75:
        medal_result.append('51-75')
    elif value > 75 and value <=100:
        medal_result.append('76-100')
    elif value > 100 and value <= 200:
        medal_result.append('101-200')
    elif value > 200 and value <= 300:
        medal_result.append('200-300')
    else:
        medal_result.append('Over 300')
```

```python
# Creating a new column to save medal tiers into
mast_df['medal_tiers'] = medal_result
```

```python
# Reviewing counts of different medal tiers
mast_df['medal_tiers'].value_counts(dropna = False)
```

```python
#Checking df
mast_df
```

### Is GDP reflected in medal groupings?

```python
sns.set(style="ticks")
g = sns.catplot(x="medal_tiers", y="GDP_value", hue="medal_tiers", data=mast_df)
plt.xticks(
    rotation=45, 
    horizontalalignment='right')
#saving visualisation
plt.savefig('GDP vs Medal Tiers scat.png')
```

### Notes on visualisation
 - The 200-300 Medal bracket is the most contradictory bracket as it has data points spread across the biggest range of GDP values. 
 - Most of the points across all medal brackets are under 25.
 - This reaffirms how little impact the GDP has on Olympic performance. 

### Is HDI reflected in medal groupings?

```python
sns.set(style="ticks")
g = sns.catplot(x="medal_tiers", y="HDI_value", hue="medal_tiers", data=mast_df)
plt.xticks(
    rotation=45, 
    horizontalalignment='right')
#saving visualisation
plt.savefig('HDI vs Medal Tiers scat.png')
```

### Notes on Visualisation
- All of the higher earning medal brackets have a high HDI value, this is consistant with previous observations
- The most diverse bracket is the 'No medal' one where you can see that countries with every HDI have received no medals, this indicates that it is not HDI alone that decides the number of medals earnt. 

### Is total population reflected in medal brackets?

```python
sns.set(style="ticks")
g = sns.catplot(x="medal_tiers", y="tot_pop", hue="medal_tiers", data=mast_df)
plt.xticks(
    rotation=45, 
    horizontalalignment='right')
#saving visualisation
plt.savefig('pair_polt.png')
#saving visualisation
plt.savefig('tot_pop vs medal tiers scat.png')
```

##### Notes on visualisation
 - While there are some extreme values in this visualisation, most of the brackets have similar population level. This emphasises that population does not have an impact on total medals

### Grouping HDI_value

```python
# Inspecting HDI_value column to assess how it should be catorgoriesd. 
mast_df['HDI_value'].describe()
```

```python
# Creating loop to catorgerise HDI rankings
HDI_result = []
for value in mast_df['HDI_value']:
    if value < 0.55:
        HDI_result.append('Low')
    elif value > 0.55 and value <= 0.70:
        HDI_result.append('Medium')
    elif value > 0.70 and value <= 0.79:
        HDI_result.append('High')
    else:
        HDI_result.append('Very high')
```

```python
# Creating a new column to save HDI_ranking into
mast_df['HDI_ranking'] = HDI_result
```

```python
# Checking value count
mast_df['HDI_ranking'].value_counts(dropna = False)
```

```python
# Checking df
mast_df
```

### Exploring HDI rankings in relation to other variables.

```python
sns.set(style="ticks")
g = sns.catplot(x="HDI_ranking", y="tot_medal", hue="HDI_ranking", data=mast_df)
#saving visualisation
plt.savefig('tot_medal vs HDI ranking scat.png')
```

```python
sns.set(style="ticks")
g = sns.catplot(x="HDI_ranking", y="GDP_value", hue="HDI_ranking", data=mast_df)
#saving visualisation
plt.savefig('GDP vs HDI ranking.png')
```

#### Note: These visualisations are consistant with the previous findings, the very high HDI rating receive the most medals and countries with higher GDP tend to have a high HDI. 

### During the cleaning process I defined the top ten performing countries. I will proceed to seperate those from the rest to find any similarities or differences. 

```python
# Creating loop to distinguish the top ten performing countries from the rest. 
top_ten = ['United States', 'Russia', ' China', 'Australia', 'Canada', 'France', 'Great Britain', 'Italy', 'South Korea']
top_ten_result = []
for value in mast_df['country']:
    if value in top_ten:
        top_ten_result.append('Top Ten')
    else:
        top_ten_result.append('Not top ten')
print(top_ten_result)
```

```python
# Saving top_ten_results into a new column
mast_df['Ranking'] = top_ten_result
```

```python
# Checking df
mast_df.head(1000)
```

### Where do the top ten sit in relation to medal brackets and HDI_value?

```python
sns.set(style="ticks")
g = sns.catplot(x="medal_tiers", y="HDI_value", hue="Ranking", data=mast_df)
plt.xticks(
    rotation=45, 
    horizontalalignment='right')
#saving visualisation
plt.savefig('HDI vs medal tiers with ranking.png')
```

### Notes on Visualisation
