## 2. Appling the elbow technique

```python
num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
# Creating a score that represents a rate of variation for the given cluster option
score = [kmeans[i].fit(df).score(df) for i in range(len(kmeans))] 

score
```

```python
# Plotting the elbow curve.

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
plt.savefig('Elbow curve.png')
pl.show()
```

#### The optimal number of clusters is 3

## 3. K-means clustering

```python
# Creating the k-means object.

kmeans = KMeans(n_clusters = 3, n_jobs = -1) 
```

```python
# Fitting the k-means object to the data.

kmeans.fit(df)
```

```python
df['clusters'] = kmeans.fit_predict(df)
```

```python
df
```

```python
df['clusters'].value_counts()
```

### 4. Creating Visualisations

```python
# Plotting results

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['HDI_value'], y=df['tot_medal'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('HDI Value') # Label x-axis.
plt.ylabel('Total Medals') # Label y-axis.
plt.savefig('Tot_Med vs HDI Cluster.png')
plt.show()
```

### Notes
 - There is not a great deal of distinction between these clusters except for the 0 (pale pink) which is the largest cluster and holds the majority of the smaller medal counts. 
 - The dark and medium purple are fairly mixed in their placement. 


```python
# Plotting results

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['GDP_value'], y=df['tot_medal'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('GDP Value') # Label x-axis.
plt.ylabel('Total Medals') # Label y-axis.
plt.savefig('Tot_Med vs GDP Cluster.png')
plt.show()
```

### Notes
- This visualisation is far more clearly defined. The clusters are split into three distinct GDP price brackets. As pale pink is by far our largest cluster it is densely grouped between 0-0.2 GDP and 0-175 medals. Whereas dark purple covers our most isolated data points, the ones with the highest GDP value and the widest range of medals. Purple is the most evenly spread between the two.
